```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('studentgrades.csv', encoding = 'latin1')
```

# DATASETS

```python
a = data['Grades']
b = data[['Hours_of_Study']]
c = data[['Attendance_Percentage']]
d = data[['Previous_Exam_Scores']]
e = data[['Extracurricular_Activities']]
f = data[['Travel_Time_to_School']]
```

# 1. DATA EXPLORATION

```python
print(data.head(50))
```

  I think every feature in this dataset contributes to the student grades. Grades will be change when some value in the dataset gets change or remove.

# 2. DATA PREPROCESSING

```python
statistics = data.describe()
print(data.isnull().sum())
```

This process enhances data analysis and modeling accuracy by mitigating the influence of varying scales on machine learning models.

# 3. TRAIN TEST SPLIT

```python
x =data.drop('Grades', axis=1)
y =data['Grades']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2,random_state=42)
```

# 4. LINEAR REGRESSION MODEL

FIT THE LINEAR REGRESSION TOOL

```python
model1 = LinearRegression()
model1.fit(b,a)
model2 = LinearRegression()
model2.fit(c,a)
model3 = LinearRegression()
model3.fit(d,a)
model4 = LinearRegression()
model4.fit(e,a)
model5 = LinearRegression()
model5.fit(f,a)
```

VISUALIZE THE LINEAR REGRESSION USING CHARTS

```python
plt.scatter(b,a)
plt.plot(b,model1.predict(b), color='red')
plt.xlabel('Hours of Study')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.show()
```

```python
plt.scatter(c,a)
plt.plot(c,model2.predict(c), color='red')
plt.xlabel('Attendance')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.show()
```

```python
plt.scatter(d,a)
plt.plot(d,model3.predict(d), color='red')
plt.xlabel('Previous Exam Scores')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.show()
```

```python
plt.scatter(e,a)
plt.plot(e,model4.predict(e), color='red')
plt.xlabel('Extra Activities')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.show()
```

```python
plt.scatter(f,a)
plt.plot(f,model5.predict(f), color='red')
plt.xlabel('Travel time to school')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.show()
```

This charts show how changes in some of this feature contributes to the dependent variable. The importance of this is to know what features in this dataset contribute to the changes of student grades.

# 5. MODEL TRAINING

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

# 6. PREDICTION AND EVALUATION

```python
new_data = np.array([[7,91,83,0,30]])
prediction = model.predict(new_data)
print(f"predicted grade: {prediction[0]}")
```

# 7. FEATURE IMPORTANCE AND ANALYSIS

The importance of each feature contribute greatly in predicting the students grade. Because every value contribute to the grade. It change when you change the value of one of the feature of it.



```python
slope = model1.coef_[0]
intercept = model1.intercept_
print('slope of hours of study: ', slope)
print('intercept of hours of study: ', intercept)
print()
slope = model2.coef_[0]
intercept = model2.intercept_
print('slope of attendance: ', slope)
print('intercept of attendance: ', intercept)
print()
slope = model3.coef_[0]
intercept = model3.intercept_
print('slope of previous exam score: ', slope)
print('intercept of previous exam score: ', intercept)
print()
slope = model4.coef_[0]
intercept = model4.intercept_
print('slope of extracurricular activities: ', slope)
print('intercept of extracurricular activities: ', intercept)
print()
slope = model5.coef_[0]
intercept = model5.intercept_
print('slope of travel time to school: ', slope)
print('intercept of travel time to school: ', intercept)
```
